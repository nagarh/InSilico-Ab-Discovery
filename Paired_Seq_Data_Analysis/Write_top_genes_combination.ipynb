```python
import pandas as pd
import os
```

```python
dir = '/data/hn533621/InSilico-Ab-Discovery/Paired_Seq_Data_Analysis/'
os.chdir(dir)
```

```python
data= pd.read_csv('Quality_checked_paired_seq.csv', low_memory=False)
```

```python
print(data['BType'].unique())
print(data['Disease'].unique())
```

```python
# Filter the data for BType 'Naive-B-Cells' and Disease nan (None)
filtered_df = data[(data['BType'] == 'Naive-B-Cells') & (data['Disease'].isna())]
```

```python
filtered_df
```

```python
# Remove the further classification of the V genes 
# Eg. IGKV1-9*01 -> IGKV1 
filtered_df.loc[:, 'v_heavy'] = filtered_df['v_call_heavy'].str.split('-', expand=True)[0]
filtered_df.loc[:, 'v_light'] = filtered_df['v_call_light'].str.split('-', expand=True)[0]
```

```python
print(filtered_df.columns)
```

```python
# From the analysis in notebook 'Sequence_Data_PreProcessing.ipynb' we know that IGHV3 and IGKV1 is the more frequent pair combination \ 
# in paired human antibody sequences
# Filter rows where 'v_heavy' is 'IGHV3' and 'v_light' is 'IGKV1'
filtered_pairs = filtered_df[
    (filtered_df['v_heavy'] == 'IGHV3') & (filtered_df['v_light'] == 'IGKV1')
]

filtered_pairs

# Save the filtered data to a CSV file
filtered_pairs.to_csv('IGHV3_IGKV1_Paired_BType_Naive_B_Cells_Disease_None.csv', index=False)

print(f"Saved {len(filtered_pairs)} rows to csv file")
```
